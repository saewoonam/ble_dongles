```python
import numpy as np
import matplotlib.pyplot as plt
import glob
```

```python
files = glob.glob("x*.csv")
files.sort()
print(files)
```

```python
def plot_rssi(rssi):
    # compute histogram fast
    hist = np.array(np.unique(rssi, return_counts=True)).T
    
    # definitions for the axes
    left, width = 0.1, 0.65
    bottom, height = 0.1, 0.65
    spacing = 0.005


    rect_scatter = [left, bottom, width, height]
    rect_histy = [left + width + spacing, bottom, 0.2, height]

    # start with a rectangular Figure
    plt.figure(figsize=(16, 8))

    ax_scatter = plt.axes(rect_scatter)
    ax_scatter.tick_params(direction='in', top=True, right=True)

    ax_histy = plt.axes(rect_histy)
    ax_histy.tick_params(direction='in', labelleft=False)

    ax_scatter.plot(rssi,'.',  markersize=1)

    ax_histy.barh(hist[:,0], hist[:,1])
    ax_histy.set_ylim(ax_scatter.get_ylim())
    return ax_scatter, ax_histy
    #plt.show()
```

```python
for filename in files:
    #filename = "5m_z_z.dat"
    data = np.genfromtxt(filename, dtype='str', delimiter=',')
    rssi = np.array(data[1:,2],dtype=np.int)
    print(filename, ' Number of stray encounters: ',(data[1:,3]!=data[1,3]).sum())
    #print(filename)
    s,h = plot_rssi(rssi);
    s.set_title(filename);
```
